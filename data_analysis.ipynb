### Importando librerías

```python
import os
import sys
import pandas as pd
from pathlib import Path
```

### Variables Globales para trabajar

```python
#BASE_DIR = Path(__file__).resolve().parent     #For productive environment
report_date = "20200801"
file_names_tamales_inc = ["Centro","E._Privados", "Norte","Sur"]
#BASE_DIR+"/"       #For productive environment
#'./teinvento_inc/ventas_reportadas_mercado_tamales/mx/'   #For testing environment
file_path_teinvento_inc = './teinvento_inc/ventas_reportadas_mercado_tamales/mx/'+report_date+'/fact_table/'
file_names_teinvento_inc = os.listdir(file_path_teinvento_inc)
file_names_teinvento_inc
```

### Definiendo Función Extract

Dentro de las primeras partes del proceso, es poder comprender como se encuentran los datos y la creación de los respectivos Data Frames a trabajar. Aquí 

```python
def extract():
    for name in file_names_tamales_inc:
        #BASE_DIR+"/"       #For productive environment
        #'./tamales_inc/ventas_mensuales_tamales_inc/mx/'    #For testing environment
        file_path_tamales_inc = './tamales_inc/ventas_mensuales_tamales_inc/mx/'+report_date+'/csv/'+name+"/ventas_mensuales_"+name+".csv"
        if name == "Centro":
            df = pd.read_csv(file_path_tamales_inc, header = None, names = ['Year','Month','Country','Calorie_type','Flavor','Zone','Resellers_id','Combo_type','Sales'])
        else:
            df_2 = pd.read_csv(file_path_tamales_inc, header = None, names = ['Year','Month','Country','Calorie_type','Flavor','Zone','Resellers_id','Combo_type','Sales'])
            df = df.append(df_2, ignore_index = True)
    df['Month'] = pd.to_datetime(df['Month'], errors = 'coerce', format = "%b")
    df['Month'] = df['Month'].dt.strftime('%m')
    df.convert_dtypes().dtypes
    return df
```

```python
df = extract()
```
